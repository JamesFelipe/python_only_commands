# Funções(def)
* São trechos de código organizados que posso utilizar quando quiser no decorer do programa
* def indica que algo é uma função
* funções podem ser com parametro ou sem parametro
  * parametro vão no corpo da função
* argumentos -> é o que vai dentro da chamada da função
* valores de retorno é o valor da chamada da função
* o comando return serve para retornar algo dentro da função

# Docstrings
* Seria algo como uma pequena documentação de funções, ou seja, o que a função faz.
* docstrings vão no início da função e são caracterizadas por seis aspas ou 12 aspas

```python
# Criando uma função, e docstring


def hello():
  """
  Isso é uma docstring
  essa função serve para exibir a mensagem olá
  """
  print('Olá')
  print()


# chamando uma função
hello()


# Função com parametro
def ola(nome):
  print('Hello', nome.title())


# chamando a função e passando seu argumento
ola('carvalho')
```

```python
# Função com retorno
def idade(i):
  return 'sua idade é ', i


id = input('Digite sua idade: ')
print(idade(id))
```

# Argumentos Nomeados(keiword arguments)
* É um par nome-valor passado para uma função.
* keiword arguments fazem com que eu não precise me preocupar com a ordem correta de seus argumentos na chamada da função e deixam claro o papel de cada valor na chamada
* são identificados pela palavra chave inserida antes dele na chamada da função
* Geralmente são parametros opcionais, por exemplo o print tem o parametro opcional end
  * o end=' ' serve para fazer com que o print continue na mesma linha ao invés de pula-la


# Valores deafult
* Podemos definir um default para cada parâmetro. Se um argumento para um parametro for especificado na chamada da função, python usará o valor desse argumento. Se não for, o valor default do parametro será utilizado.
* Ao usar valores default, qualquer parametro dese tipo deverá ser listado após todos os parâmetros que não tenham valores default. Isso permite que python continue a interpretar argumentos posicionais corretamente
* Resumindo, ele vai no corpo da função

# Escopo local e global
* Escopo local: são variáveis e os parametros que alguma função recebe, elas são exclusivas das funções a que pertecem, ou seja, esse tipo de variável só pode ser usado dentro dessa mesma função

* escopo global: valores recebidos fora das funções, podem ser usadas durante todo o programa, até fora da função que foi criado

* variáveis locais não podem ser usadas em um escopo global, nem mesmo outras variáveis locais podem ser usadas em escopo local
* variáveis globais podem ser lidas a partir de um escopo local
* é permitido nomes iguais com variavéis do escopo global e local, mas é uma ideia ruim, assim como usar variáveis globais também é uma ideia ruim
* caso seja necessário utilizar variáveis globais utilize o comando global, assim, nenhuma variável local de mesmo nome será criada

# Funções com retorno
* Basicamente são dados processados, análisados e retornados, podendo ser desde dados básicos a dados mais complexos.
* return toma um valor que está em uma função e o envia de volta a linha que o chamou
* funções com retorno não costumam ter valor None, o tipo delas dependem dos tipos básicos


```python
def soma(a, b):
  """
  Essa função os valores a + b e retorna o resultado em c
  """
  c = a + b
  return c


print(help(soma))
print(soma(1, 1))
print(type(soma(13, 1)))
```

```python
# Argumento posiconal
def soma2(a, b=0):
  c = a + b
  return c


print(soma2(13))
print(soma2(13, 12))
```

```python
# Devolvendo um dicionário
def info_pessoa(first_name, last_name):
  pessoa = {'first': first_name, 'last': last_name}
  return pessoa


print(info_pessoa('james', 'felipe'))
```

```python
# Função com o laço while
def info_pessoa(first_name, last_name):
  pessoa = {'first': first_name, 'last': last_name}
  nome_completo = first_name + ' ' + last_name
  return nome_completo.title()
  
# loop infinito
while True:
  f_name = input('Digite seu primeiro nome: ')
  l_name = input('Digite seu último nome: ')
  print('Olá, ', info_pessoa(f_name, l_name))
  opcao = input('Você quer continuar s/n: ').strip().lower()
  if opcao == 'n':
    break
```

```python
# passando uma lista para a função
def usuarios(nomes):
  for nome in nomes:
    print('Olá, ', nome.title())


usernames = ['margoth','azatoth', 'azazel']
print(usuarios(usernames))
```

# args( número arbitrário de argumentos)
* Assim sendo, dá para passar quantos argumentos eu quiser para a função
* é indicada por um asteristico
* a palavra args é um padrão da linguagem, mas podia ser qualquer outro nome

# kwargs(argumentos nomeados arbitrários)
* assim sendo, dá para passar quantos argumentos eu quiser sendo valores do tipo chave valor
* É indicada por dois asteristicos
* a palavra kwargs é um padrão da linguagem, mas podia ser qualquer outro nome

```python
def cidades(*args):
    return args

print(cidades('irituia', 's. miguel do guamá', 'belém'))
```

```python
# livro curso intensivo de pyton página 189
def dados_user(first, last, **user_info):
  """
  constroí um dicionário contendo tudo que sabemos sobre um usuário
  """
  profile = {}
  profile['first_name'] = first
  profile['last_name'] = last
  for key, value in user_info.items():
    profile[key] = value
    return profile
  

user_profile = dados_user('albert', 'eistein', location='princeton', 
                          field='phisics')
print(user_profile)
```
