# Módulo logging
* Serve para rastrear erros, aonde criamos o código e o logging cria um arquivo txt, e daí podemos rastrear erros

# Níveis de Logging
* Nível &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Valor númerico
* CRITICAL&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 50
* Error&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 40
* WARNING&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 30
* INFO&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 20
* DEBUG&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 10
* NOTSET&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 0

# Comandos
* o comando log.nivel() -> serve para registrar eventos de cada um desses níveis
* basicConfig(** kwargs) -> serve para fazer as configurações básicas do logging
  * parametros
    * level: O logger raiz será definido para o nível de loggin especificado.
    * filename: especifica o nome do arquivo.
    * filemode: Se o nome do arquivo for fornecido, o arquivo é aberto neste modo. O padrão é a , o que significa anexar.
    * format: este é o formato da mensagem de registro.


```python
import logging

# Abaixo nome do arquivo aonde os códigos serão registrados
# Por padrão, não preciso colocar o level, mas pode ser necessário
logging.basicConfig(filename="app.log", level=logging.DEBUG)

# É necessário criar alguma coisa, para realmente criar o arquivo
log = logging.getLogger()

# Utilizando tipo de registro
log.info("Olá")
log.critical('Erro grave')
log.level
```

```python
# usando o level para definir o nível da mensagem
import logging

logging.basicConfig(level=logging.DEBUG)
logging.debug('Isso será registrado')
# assim, todos os níveis debug ou acima serão registrados
```
