# While(enquanto)
* Nessa caso o bloco de código será executado enquanto a instução for verdadeira
  * as vezes é necessário um flag(ponto de parada)
  * cuidado, pois se esquecer de algo no while, ou se passar uma condição sem ser lógica podemos ir para um loop infinito, ou seja, o programa executará eternamente


```
while condição:
  <bloco de código>
```


# break(parada)
* serve para forçar a parada de um programa

# continue
* serve para pular um valor especifico



`````python
n = 1
while n < 5:
  print(n)
  n += 1
`````

`````python
# break
cidades = []
while True: # loop infinito
  city = input('Digite uma cidade: ').title()
  cidades.append(city)
  op = str(input('Você quer continuar? Sim/Nao ')).lower().strip()
  if op == 'nao':
    break

print('Cidades: ', cidades)
`````

`````python
# Nomes de gatos
gatoNomes = []

while True:
  nome = input('Digite o nome do {}º gato: '.format(len(gatoNomes)+1))
  if nome == '':
    break
  gatoNomes = gatoNomes + [nome] # concatenação
  
print('O nome dos gatos são: ')
for nome in gatoNomes:
  print(nome)
`````

`````python
# continue
m = 0
while m < 9:
  m += 1
  if m == 3:
    continue
  print(m)

    
   
`````

# Função range()
* serve para criar uma sequência de valores, podendo ser, apenas um valor, ou início até o final e, e início até o final com passo
* começa em zero e não em 1
* também aceita valores negativos

`````python
# for
numeros = [1, 2, 3, 1, 5, 6, 7]
for numero in numeros:
  print(numero, end=' ')
print()

# for com contador
contador = 0
for num in numeros:
  contador += 1
print(f'Temos {contador} números na lista')

# Somando todos os valores com for
contador = 0
for num in numeros:
  contador += num
print(f'Temos {contador} total')

# função range
print(range(10))
print()

# for com range
for n in range(0, 10):
  print(n, end=' ')
`````

# Sys.exit()
* serve para encerrar o programa no momemto que quisermos

`````python
import sys
while True:
  print('Digite exit para sair.')
  resposta = input()
  if resposta == 'exit':
    sys.exit('Você digitou ' + resposta + '.')
`````

`````python
"""
Crie um pequeno programa que mostre os números de 1 a 10 usando um loop for.
Em seguida, crie um programa equivalente que mostre os números de 1 a 10
em um loop while
"""
for n in range(1, 11):
  print(n, end=' ')
print()

n = 1
while n <= 10:
  print(n, end=' ')
  n += 1
`````
