# Arquivos
* Python usa objetos de arquivo para interagir com arquivos externos em seu computador. Podendo ser de qualquer tipo, porém o mais usado costuma ser o texto

* %%writefile nome_do_arquivo.txt -> serve para criar um arquivo de texto(ipython), sendo que o texto vai abaixo do comando 

# Python Comandos
* open('arquivo.txt') -> serve para abrir um arquivo
* arquivo.read() -> depois de aberto, o read serve para realmente ler o arquivo
* arquivo.seek(0) -> serve para voltar o curso de leitura para o início do arquivo
* arquivo.readlines() -> lê o arquivo linha a linha, mas cuidado, pois fica tudo salvo na memória, arquivos pequenos são melhores para esse método
  * realines -> retorna uma listas das linhas no arquivo
* open(open('arquivo.txt', 'w') -> o 'w' abre o arquivo para escrita
* arquivo.write('texto') -> escreve no arquivo, cuidado, pois pode acabar sobreescrevendo valores, use um \n antes de escrever para pular para a próxima linha
* arquivo.close() -> serve para fechar o arquivo.


```python
# criando arquivo de texto com o ipython
%%writefile test.txt
Hello, this is a quick test file

```

```python
# Abrindo arquivo 
arquivo = open('test.txt')
print(arquivo)

# Lendo arquivo(curso vai do início ao fim)
print(arquivo.read())

# Lendo o arquivo de novo(o curso está no final do arquivo)
print(arquivo.read())

# redefiindo o curso ao ponto inicial
print(arquivo.seek(0))

# Mesmo processo de leitura
print(arquivo.read())

# readlines()
print(arquivo.readlines())


```

```python
# Abrindo para escrita
arquivo = open('test.txt', 'w+')

# Escrevendo no arquivo
arquivo.write('I program in Python')


arquivo.seek(0)
# lendo
print(arquivo.read())
```

```python
james = open('arquivo.txt', 'w+')
james.write('james')
print(james.read())
james.write('\nJames Felipe')
james.seek(0)
print(james.read())
print(james.close())
```

# Map
* map(função, coleção) -> aplica a função a todos os
elementos da seqüência. Ele retorna uma nova lista com os elementos alterados por função
* para de fato vizualizar valores depois de usar o map é necessário converte seus valores para lista(list)
* o map serve para economizar linhas e costuma ser usado mais com lambda

```python
# Elevando valores
a = map(lambda x: x**2, [1, 2, 3, 4, 5])

# Exibindo o valor de map
print(a)

# Exibindo os valores em uma lista
print(list(a))

```

```python
# Função e depois map
def soma(a, b):
  c = a + b
  return str(c)

s = map(soma, [1, 2], [1, 2])
print(list(s))
```
