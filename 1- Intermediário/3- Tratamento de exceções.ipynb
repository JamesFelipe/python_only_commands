# Tratamento de execeções
* Erros são exceções, o que faz com que o programa todo falhe, mas o tratamento de execções serve para fazer com que esses eros não causem problemas mais sérios
* dentro do try temos o código que pode conter erros e no except vai o erro
* finally sempre será exibido
* Sendo que quando um erro acontecer, irá diretamente ao except para ser exibido a mensagem de erro que eu selecionar

```python
# Exceção: não podemos dividir por zero
def divisao(dividePor):
  try:
    return 42 / dividePor
  except ZeroDivisionError:
    return 'Não podemos dividir por zero'


print(divisao(2))
print(divisao(12))
print(divisao(0))
```

```python
# código para saber o erro
def divisao(a, b):
  try:
    print(a/b)
  except Exception as e:
    print('Divisão inválida')
    print('Erro: ', e)
  finally:
    print('fim')


divisao(2, 0)
```

# Apresentação elegante
* São melhores formas do print e do format, apresentações mais limpa
  * é necessário importar
* pretty print == pprint()

```python
import pprint
msg = 'Hoje é um dia comum de abril, e iremos fechar na terça'
conta = {}

for caractere in msg:
  conta.setdefault(caractere, 0)
  conta[caractere] = conta[caractere] + 1

pprint.pprint(conta)
```
