# Funções Lambda
* São Funções anononimas, ssão caracterizadas pela palavra lambda e ter apenas costumam ter uma linha

# map()
* calcula todos os valores de um iteravel
* sintaxe: map(função, iteravel)
  * o iteravel pode ser coleção
* É necessário converter o valor para uma coleção para que posso vizualizar os dados

# filter()
* parecido ao map, filtra os valores para que seja possível realizar operações

```python
a = lambda x: x**2
a(2)
 
```

```python
seq = [1, 2, 3, 4, 5]
valores = map(lambda x: x**2, seq)
list(valores)
```

```python
# Retornado valores impares
valores2 = filter(lambda x:x % 2==1, [1, 2, 3, 4, 5, 10])
list(valores2)
```
