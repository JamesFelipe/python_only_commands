# Ambientes virtuais
* É como se fosse um novo "computador" com suas pastas e principalmente o python
* Ambientes virtuais servem para fazer testes sem que causem erros por toda a máquina, portanto, caso tenha um erro grande, bastaria se desfazer da máquina virtual
* Apesar de existir pacotes especificos para criar ambientes virtuais, o 'venv' costuma já vim pré-instalado
