# List Comprehensions(Lista inteligente)
* É uma lista compactada, ou seja, dá para usar como uma única linha de código
* sintáxe -> [saída: coleção: condição]
    * parte impressa: for: condição
    * parte impressa(expressão): for: range ou condição

```python
squares = [value**2 for value in range(1,11)]
squares
```

```python
# Valores Pares
[x for x in range(5) if x % 2 == 0]
```
