# Herança
* Herda valores, como atributos e métodos da classe-pai(também chamada classe superior ou classe ancestral, etc)
* Outra maneira de ver é herança é uma forma de formar novas classes usando classes que já foram definidas.
* a classe-pai vai dentro de um parentesis na classe filha, para de fato herdar o que tem
* classes filhas podem inclusive sobreescrever métodos da classe pai
* Use isinstance() para verificar o tipo de uma instância: isinstance(obj, int) será True somente se obj.__class__ é a classe int ou alguma classe derivada de int.
* Use issubclass() para verificar herança entre classes: issubclass(bool, int) é True porque bool é uma subclasse de int. Porém, issubclass(float, int) é False porque float não é uma subclasse de int.

````python
# o filho busca a herança kkkk
class Animal:
  def __init__(self):
    print('Animal criado!')


  def tipo(self):
    print('Animal')

  def comendo(self):
    """
    Esse método representa um animal comendo
    """
    print('Comendo')
  



# agora sim está herdando de fato
class Gato(Animal):
  def __init__(self):
    print('Gato Criado')

  def tipo(self):
    print('Gato')

  def onomatopeia(self):
    print('Miau')



# Trabalhando com a classe herdada e com a classe pai
ga = Gato()
ani = Animal()
print()

# tipo 
ga.tipo()
ani.tipo()

# novo método criado
ga.onomatopeia()
print()

# método não criado da classe-pai na classe filha
ga.comendo()
````

````python
# Documentação python(variáveis privadas)
class Mapeamento:
    def __init__(self, iterable):
        self.items_list = []
        self.__update(iterable)

    def update(self, iterable):
        for item in iterable:
            self.items_list.append(item)

    __update = update   # cópia privada do método update() original

class MapeamentoFilho(Mapeamento):

    def update(self, keys, values):
        # fornece uma nova assinatura para update()
        # mas não quebra __init__()
        for item in zip(keys, values):
            self.items_list.append(item)


mapa = Mapeamento([1, 2, 3])
print(mapa.items_list)
mapa2 = MapeamentoFilho({'nome': 'Noname', 'idade': 0})
mapa2.update(['nome', 'idade'], ['Noname', 0])
print(mapa2.items_list)
````

# Alguns métodos especiais
* _ _ init _ _(self) -> inicia  a classe
* _ _ str _ _(self) -> é usada para retornar uma representação de string de um objeto
* _ _ len _ _(self) -> é usada para fazer contagem de elementos e retornará um valor inteiro
* _ _ del _ _(self) -> é um método destruidor que é chamado assim que todas as referências do objeto são excluídas


````python
class Livro(object):
  def __init__(self, titulo, autor, paginas):
    print('Livro Criado')
    self.titulo = titulo
    self.autor = autor
    self.paginas = paginas

  def __str__(self):
    return "Título:%s, Autor:%s, Páginas%s"%(self.titulo, self.autor, self.paginas)

  def __len__(self):
    return self.paginas

  def __del__(self):
    print('Livro Destruído')




livro = Livro("Curso intensivo de python", "Erick Matheus", 592)
livro.__str__()
livro.__del__()
````

# Iteradores
* Um iterador é sempre um iterável, mas que produz um valor a cada vez que é usado como argumento da função nativa next()
* python chama a função iter() que retorna um objeto que define next()

````python
# Código documentação python
class Reverso:
    """Iterador para repetir uma sequência para trás."""
    def __init__(self, data):
        self.data = data
        self.index = len(data)

    def __iter__(self):
        return self

    def __next__(self):
        if self.index == 0:
            raise StopIteration
        self.index = self.index - 1
        return self.data[self.index]



reverso = Reverso('valor')
print(reverso.__iter__())
reverso.__next__()

for cha in reverso:
  print(cha)
````

# Geradores
* Servem para criar iteradores
* são escritos como funções normais mas usam a instrução yield quando precisam retornar dados.

# Expressões Geradoras
* São geradores simples, escritos parecidos com list comprehension, porém com parentesis ao invés de colchetes
* usam menos mémoria
* são mais compactadas que geradores normais
* e são projetadas para serem usadas imediatamente


````python
# código documentação python
def reverse(data):
    for index in range(len(data)-1, -1, -1):
        yield data[index]

for cha in reverse('Brasil'):
  print(cha, end=' ')
````


# Decoradores
* função que atribui uma nova funcionalidade para outra função.
* crio a função e posso ou não usar o @ para definir um decorador

````python
# criando uma nova função
def func():
  return 1

print(func())  
# Funções Globais que posso usar
print(globals())

# trasnformando em chaves de dicionário
print(globals().keys())
````

````python
# Funções locais
def func():
  print(locals())

func()

# Criando decoradores
def novo_decorador(func):

  def executa_func():
    print('Executando a função func')
    func()
    print('Código executado depois de func')

  return executa_func


def func_precisa_decorador():
  print('Essa função precisa de um decorador')



# chamando função
func_precisa_decorador()

# redefinindo a função
func_precisa_decorador = novo_decorador(func_precisa_decorador)
print()

# chamando novamente a função
func_precisa_decorador()
print()

# Usandoa notação @
@novo_decorador
def func_precisa_decorador2():
  print('Precisando decorador')


# Exibindo
func_precisa_decorador2()  
````
