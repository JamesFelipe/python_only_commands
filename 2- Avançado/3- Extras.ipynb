```python
# Fibonnaci(código documentação python modificado)
a, b = (0, 1)
valor = int(input('Digite até que valor você quer ver a sequência: '))
while a < valor:
  print(a, end=' ')
  a, b = b, a + b
```

# Filas
* Filas são estruturas de dados em que só é possível inserir um novo elemento no final da fila e só é possível remover um elemento do início. 
  * filas seguem um protocolo em que o primeiro a entrar é o primeiro a sair.
* é necessário importar o comando deque da biblioteca collections

```python
from collections import deque
lista = deque(['James', 'Felipe', 'Ananda'])
print(type(lista))
print(lista)
lista.append('Fernanda')
print(lista)
lista.append('Jacó')
print(lista)
print(lista.popleft())# removendo o primeiro valor do lado esquerdo
print(lista)
```

# Algumas Funções
* enumarate(coleção) -> Devolve tuplas com seus respectivos valores disponíveis
* zip(coleção1, coleção2) -> Funciona quase como o enumarete, mas posso passar duas coleções

```python
for i in enumerate([1, 2, 3, 4]):
  print(i)
```

```python
for i, n in zip([1, 2, 3], [4, 5, 6]):
  print(f'1: {i} 2:{n}')
```
