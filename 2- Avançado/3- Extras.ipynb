```python
# Fibonnaci(código documentação python modificado)
a, b = (0, 1)
valor = int(input('Digite até que valor você quer ver a sequência: '))
while a < valor:
  print(a, end=' ')
  a, b = b, a + b
```

```python
# Passando um único valor para várias variáveis
v1 = v2 = v3 = 'James'
print(v1, v2, v3)
print()

# Desempacotamento com tuplas
frutas = ('manga', 'jaca', 'amianto') 
(ruim, boa, suco) = frutas
print(boa, ruim, suco)

# Desempacotando com asteristico
(boa, *ruim) = frutas
print(boa, ruim)
```
