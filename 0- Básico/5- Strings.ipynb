# Strings e seus métodos
* string.capitalize() -> deixa a letra inicial do primeiro caractere maiúscula
* string.casefold() -> faz com que toda os caracteres fiquem minúsculos
* string.center(valor_de_centralização) -> serve para centralizar os caracteres,podendo ter apenas o valor, ou pode ter também os espaços  a serem preenchidos
* string.count('valor') -> busca valroes, e retorna a quantidade que encontrar
* string.find('valor') -> busca a primeira ocorrência e retorna esse valor, temos que ter cuidado, pois ele inicia a contagem em zero
* format -> serve para melhorar a visualização do print
  * print('{} + {} = {}'.format(num1, num2, num1 + num2))
* index -> é um find com um nome diferente kkkk, ele busca um determinado valor e retorna a primeira posição que encontrar, a contagem dele se inicia em zero
* ' '.join([valores]) -> união de strings, porém é necessário usar uma nova variável para realmente substituir valores
* string.lower() -> todos os caracteres em minúsculo
* string.lstrip() -> limpa os espaços em branco da esquerda
* string.replace(valor_velho, valor_novo) -> troca caracteres por outros, também é necessário uma nova variável para subsituir de fato
* string.rfind('valor') -> faz busca na direita
* string.rsplit(' ') -> divide em listas a direita, e dentro do parentesis, vai o valor de divisão, podendo ser espaços em branco, vírgulas, etc
* string.rstrip() -> limpar os espaços em branco da direita
* string.split('') -> transforma os caracteres em listas baseado no espaço em branco para ser a divisão, mas pode ser outros caracteres também
* string.strip() -> limpar os espaços em branco dos dois lados
* string.title() -> capitaliza todos os caracteres
* string.upper() -> todos os caracteres em maiúsculo

# Strings e seus métodos booleanos
* string.isalnum() -> só tem números?
* string.isalpha() -> só tem valores alfabéticos
* string.isascii() -> tem valores da tabela ascii
* string.isdecimal() -> só tem valores decimais
* string.isdigit() -> é um digito?
* string.islower() -> está todo em minúsculo?
* string.isnumeric() -> números?
* string.isspace() -> só tem espaços?
* string.istitle() -> está capitalizado?
* string.isupper() -> está todo em maiúsculo?

```python
# casefold
a = 'james felipe Magalhaes'
print(a.casefold())
print()

# capitalize
print(a.capitalize())
print(a.title())
print()

# center
print(a.center(50))
print(a.center(50, 'A'))
```

# Fatiamento(slice) de strings
* É pegar uma parte da string
  * síntaxe básica: string[inicio: fim: passo]
  * sendo que nenhum deles são obrigátorios

```python
frase = 'A jabutica é uma fruta'
print(frase[0])# primeiro caractere
print(frase[0:]) # do primeiro caractere até o final
print(frase[::]) # do primeiro caractere até o final
print(frase[::3]) # do primeiro caractere até o final indo de 3 em 3
print(frase[-1]) # último caractere
print(frase[::-1]) # invertendo a frase
```

# Função len
* retorna a quantidade de caracteres

```python
valor = 'O anel do rei é prata'
print('anel' in valor)

print(len(valor))
print(valor.lower())
print(valor.upper())
print(valor.split())
```

# Concatenação e multiplicação de strings
* (+) serve para unir duas strings
* (*) serve para multiplicar strings
  * obs.: é uma string mutiplicado por um número, não por outra strings
* len(valor) -> serve para exibir o tamanho dos caracteres, e conta os espaços juntos



```python
# concatenar
nome = 'brasil'
print(nome.title() + ' ' + 'é um páis lindo')
print()

# Multiplicando
print(nome * 10)
print(nome * len(nome))
```
