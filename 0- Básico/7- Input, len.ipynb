# Input
* comando para entrada de dados pelo teclado
* Por padrão ela é string, mas posso selecionar o seu tipo(casting)

```python
print('Digite seu nome: ')
nome = input()
print('Seu nome é ' + ' ' + nome)
```

```python
# casting direto no input
idade = int(input('Digite seu nome: '))
print('Sua idade é {}'.format(idade))
```

# Len
* Essa função conta os caracteres


```python
nome = 'james'
print(len(nome))
```
