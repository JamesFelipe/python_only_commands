# Explicações
* Algoritmo -> Sequência de passos organizados que servem para resolver problemas

* python -> é uma linguagem de programação
* arquivo.py -> o ".py" indica que tra-tase de um arquiv0o python

`` `Monospace` `` | `Monospace`(acento gravess em ambos os lados)
`~~palavra riscada~~` | ~~palavra riscada~~ (tios em ambos os lados

# Formulários no notebook
* basta usar: #@title String fields 
  * usada na primeira linha de célula de código
* #@title Date fields -> seria a entrada de datas
* #@title Number fields -> formulário de números
* #@title Boolean fields -> formulários booleanos

````python
#@title String fields
texto = 'Valor em texto' #@param {type:"string"}
opcoes = '3rd option' #@param ["1st option", "2nd option", "3rd option"]
texto_e_opcoes = 'valor' #@param ["1st option", "2nd option", "3rd option"] {allow-input: true}


print(texto)
print(opcoes)
print(texto_e_opcoes)
````

````python
#@title Date fields
data_entrada = '2022-12-11' #@param {type:"date"}

print(data_entrada)
````

````python
#@title Number fields
entrada_numerica = 10.0 #@param {type:"number"}
numero_deslizante = 0.4 #@param {type:"slider", min:-1, max:1, step:0.1}

entrada_inteira = 10 #@param {type:"integer"}
inteiro_deslizante = 83 #@param {type:"slider", min:0, max:100, step:1}

print(entrada_numerica)
print(numero_deslizante)

print(entrada_inteira)
print(numero_deslizante)
````

````python
#@title Boolean fields
booleano_checkbox = True #@param {type:"boolean"}
booleano_dropdown = True #@param ["False", "True"] {type:"raw"}

print(booleano_checkbox)
print(booleano_dropdown)
````
