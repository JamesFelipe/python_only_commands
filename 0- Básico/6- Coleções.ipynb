#  Lista(list) [ ]
* tipada, ou seja, cada elemento pode ser de um tipo diferente
* aceita slicing
* acesso elementos da lista por índexes
  * a contagem se inicia em zero não em um
* listas aceitam buscas por indexação negativa
* listas com pelo menos um elemento retorna True e listas vazias retornam False

# Métodos da lista
* lista.append(novo_valor_a_ser_adicionado) -> adiciona um valor no final da lista
* lista.clear() -> faz com que a lista fique vazia
* lista.copy() -> cria uma cópia da lista
* lista.count(valor) -> procura um elemento na lista e retorna quantas vezes o número aparece
* lista.extend([valores]) -> adciona vários novos valores a lista
* lista.index(valor) -> serve para encontrar um valor e retorna seu índice, sendo que se existe valores duplicados, o primeiro valor será exibido
* lista.insert(índice, valor_a_ser_inserido) -> adiciona valores na posição que eu selecionar a lista
* lista.pop() -> remove um valor da lista e o retorna, por padrão retorna o último valor, mas dentro parêntesis eu posso especificar esse valor
* lista.remove('elemento') -> excluir valores da lista baseado no elemento
* lista.sort() -> ordena os valores da lista baseados na tabela ascii
  * parametro **reverse=True** serve para inverte a ordenação
  * parametro **key = str.lower)** faz com que todos os caracteres sejam tratados como minúsculos
* del lista[valor] -> serve para remover tanto um elemento, quanto vários elementos da lista, se baseia no índice
* min(lista) -> pega o valor menor de uma lista
* max(lista) -> pega o valor maior de uma lista
* sum(lista) -> essa função soma todos os valores da lista

```python
# index
lista1 = ['Brasil', 'Eua', 'Italia', 'Bélgica', 'Noruega']
print(lista1.index('Italia'))


# ordenando com o sort
lista1.sort()
print(lista1)

# parametro reverse do sort
lista1.sort(reverse=True)
print(lista1)

# parametro key=str.lower
lista2 = ['Ana', 'ama', 'o', 'Rato', 'sexta', 'Sexta', 'Ama', 'O', 'Rato']
lista2.sort(key=str.lower)
print(lista2)
```

```python
# Cópia da lista
lista = [1, 2, 3, 4, 5]
lista2 = lista.copy()
print(len(lista))

lista.append(1)
print(lista)

# concatenação de lista
print(lista2)
nova_list = lista + lista2
print(nova_list)

# insert
nova_list.insert(2, 7)

# removendo
nova_list.remove(3)
print(nova_list)
print(len(nova_list))


# removendo ocm o del
del nova_list[1]
print(nova_list)
```

```python
# slice
nova_list
print(nova_list[0:2])

# indexação negativa
print(nova_list[-2])
```

```python
lista = [1, 2, 3, 4, 5]
print('Minímo: ', min(lista))
print('Máximo: ', max(lista))
print('Soma: ', sum(lista))
```

# Tuplas(tuple) ( , )
* São imútaveis
* a vírgula define a tupla, não os parentesis
* aceita, slicing de strings
* aceita indexação negativa
* acesso por índices
* Tuplas são imútaveis, mas posso sobreescrever

# Métodos da tupla
* tupla.count('valor') -> busca e retorna quantas vezes o valor aparece
* tupla.index('valor') -> faz a busca e retorna o índice da primeira vez que o valor apareceu


```python
tupla = (41, 5, 5)

# acessando
print(tupla[0])

print(tupla[-2])

print(tupla[0:1])
```

```python
# Sobreescrevendo
tupla1 = (200, 40)
print(tupla1)
tupla1 = (400, 100)
print(tupla1)
```

# Dicionário(dict) { }
* é chave: valor
* não tem ordenação
* seus elementos podem ser de qualquer tipo, porém suas chaves podem ser dos tipos básicos ou tupla, já que não podem ser alteradas
* Para acessar o dicionário utilizo as chaves dentro de colchetes

# Métodos do dicionário
* dicionario.clear() -> Faz o dicionário ficar vazio
* dicionario.copy() -> cria uma cópia do dicionário
* dicionario.get('valor', 'mensagem personalizada') -> faz buscas no dicionário e se não achar posso personalizar uma mensagem
* dicionario.items() -> retona as chaves e os valores do dicionário
* dicionario.keys() -> retorna as chaves do dicionário
* dicionario.pop('valor' -> remove valores do dicionário e o retorna o valor
* dicionario.update({chave:valor}) -> adiciona novos valores ao dicionário
* dicionario.values() -> retorna os valores do dicionário
* dicionario.setdefault('chave_existente_ou_não') -> serve para definir um novo valor ao dicionário, o 1º valor é  a chave, e 2º valor é o elemento, caso não exista será criado o novo item, se já existir o método o retornará
* sorted(dicionario) -> ordena as chaves do dicionário

```python
pessoa = {'nome': 'James None', 
          'idade': 22, 
          'Cursos': ['Pyhton', 'Html', 'SQL']}
          
print(len(pessoa))

# acessando
print(pessoa['Cursos'])

# chaves
print(pessoa.keys())

# items
print(pessoa.items())

# get
print(pessoa.get('azul', 'valor não encontrado'))

# update
print(pessoa)
pessoa.update({'País': 'Brasil'})
print(pessoa)

# clear
pessoa.clear()
print(pessoa)
```

```python
pais = {'Brasil': 210, 'Argetina': 45, 'china': '1.3B'}
print(pais)
pais.setdefault('india', '1.1B')
print(pais)

print(sorted(pais))
```

# Conjuntos(set) { }
* Possuem elementos únicos
* Não são indexados
* Não ordenado
* Todos os elementos precisam ser imútaveis
* Possui diversos métodos baseados em lógica matemática

# Algumas funções do conjunto
* set.add(valor) -> adiciona um novo valor ao conjunto
* set.clear() -> esvazia o conjunto
* set.copy() -> cria uma cópia do conjunto
* set.pop() -> remove um valor do conjunto e o retorna, indo do início até o final
* set.remove('valor') -> remove um valor especifico do set
* set.discard('valor') -> Remove o item especificado, porém se o item não existir retornará None
* set.update({valores}) -> Atualize o conjunto com a união deste conjunto e outros
* set.isdisjoint(set2) Retorna se dois conjuntos têm uma interseção ou não
* set.issubset(set2) Retorna se outro conjunto contém este conjunto ou não
* set.issuperset(set2) Retorna se este conjunto contém outro conjunto ou não

# Métodos matemáticos
* set.difference(set2) -> Retorna um conjunto contendo a diferença entre dois ou mais conjuntos
* set.difference_update(set2) -> Remove os itens neste conjunto que também estão incluídos em outro conjunto especificado
* set.intersection(set2) -> Retorna um conjunto, que é a interseção de dois outros conjuntos
* set.intersection_update(set2) -> Remove os itens neste conjunto que não estão presentes em outro(s) conjunto(s) especificado(s)
* set.symmetric_difference(set2) -> Retorna um conjunto com as diferenças simétricas de dois conjuntos
* set.symmetric_difference_update(set2) -> insere as diferenças simétricas deste conjunto e de outro
* set.union(set2) -> retorna um novo set com a união dos elementos


# Operadores binários
* | -> esse é o union, união de conjuntos
* & -> intersection, intersecção entre os conjuntos
* -(menos) -> difference, diferença entre os conjuntos(elementos do primeiro que não está no segundo)
* ^ -> symmetric_difference, diferença simetrica(o inverso da intersecção, apenas os elementos que estão somente no primeiro ou somento no segundo conjunto

# Operadores relacionais
* <=issubset -> verdadeiro se o primeiro conjunto estiver contido no segundo
* (>=)(sem parentesis)issuperset -> verdadeiro se o primeiro conjunto estiver contido no segundo


# Operadores de atribuição
* |= update -> atribui a união dos conjuntos(antes e depois do operador) a variável antes do operador
* &= intersection_update  -> atribui a intersecção dos conjuntos a variável antes do operador
* -= difference_update -> atribui a diferença dos conjuntos a variável antes do operador
* ^= symmetric_difference_update -> atribui a diferença simétrica dos conjuntos a variável antes do operador

```python
a = {1, 2, 3, 4, 5}
print(type(a))

# add
a.add(6)
print(a)

# criando uma cópia do conjunto
b = a.copy()
print(a)
print(b)
b.add(9)
print(b)
print(a)
```

```python
# Convertendo tupla para lista e vice versa
lis = list(range(1, 5))
print(lis)
print(type(lis))


tupla = tuple(lis)
print(tupla)
print(type(tupla))
```

# Informações aninhadas
* Nada mais é do que colocar alguma coisa dentro da outra

```python
# dicionários na lista
pessoa1 = {'nome': 'ana', 'idade': 18, 'profisão': 'modelo'}
pessoa2 = {'nome': 'cauã', 'idade': 18, 'profissão': 'mangaká'}
pessoa3 = {'nome': 'Felipe', 'idade': 22, 'Profissão': 'programador'}

pessoas = [pessoa1, pessoa2, pessoa3]
for pessoa in pessoas:
  print(pessoa)
print()


# lista em dicionário
favorite_languages = {
  'caio': ['c', 'python'],
  'Elon': ['c', 'python', 'c++'],
  'james': ['python', 'r', 'javasscript'],
}

for language in favorite_languages.items():
  print(language)
print()

# Dicionário em um dicionário
pessoas2 = {
    'Icáro de Carvalho': {
        'profissão': 'marketing digital'
    },

    'Thiago Finch': {
        'Profisão': 'Marketing Digital'
    },

    'Thiago nigro': {
        'Profisão': 'Marketing Digital'
    },
}

for pessoa in pessoas2.items():
  print(pessoa)
```
