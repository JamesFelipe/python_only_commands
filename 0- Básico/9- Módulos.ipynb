# Importando módulos
* import -> serve para importar funções(métodos) que eu criei ou não
* from -> serve para importar uma função especifica do módulo
  * o que aconte é que com o import toda a biblioteca é importada, já com o from apenas um valor especifico é importado, não sendo necessário mais usar o nome do módulo

```python
# biblioteca de aleátorios
import random

for i in range(1, 10):
  print(random.randint(1, 5), end=' ')


# o randint gera uma sequência de números aleátorios inteiros 
```

```python
# biblioteca mátematica e raiz quadrada
from math import sqrt
print(sqrt(81))

# importando toda a biblioteca com o from
from random import *
```

```python
"""
Escreva um código que exiba Hello se 1 estiver armazenado em spam
Howdy se 2 estiver armazenado em spam e
Greentis se outro valor estiver armazenado em spam
"""
# Definição de valor Principal(O que estiver abaixo será executado primeiro)
if __name__ == "__main__":
  # Código
  spam = 1
  if spam == 1:
    print('Hello')
  elif spam == 2:
    print('Howdy')
  else:
    print('Greentis')
```

# Cópias
* copy.copy() cria uma duplicata de um valor mútavel como uma lista e outras coleções e não apenas faz uma cópia de referência
* copy.deepcoy() copiará listas internas também além dos outros valores

```python
import copy
lis = list(range(10))
print(lis)

# copy.copy()
lis2 = copy.copy(lis)
print(lis2)
lis2.append(1)
print(lis2)

print()

# copy.deepcopy()
lis3 = copy.deepcopy(lis2)
lis3.append(5)
print(lis3)
```
