# Convensão de tipos
* int() -> converte valores para inteiro(números, não funciona com caracteres)
* float() -> converte valores para float(números, não funciona com caracteres)
* str() -> converte valores para string

```python
a = 2
b  = '2'

# inteiro para string
print(str(a) + b)

# string para inteiro
print(int(b) + a)

# inteiro para float
print(float(a))

# float para string
print(str(5.5))

# entre outros
```

# Alguns tipos de dados númericos
*  complex aceita valores complexos
* Decimal aceita valores decimais, e precisa ser importado de uma biblioteca especifica

```python
# complexo
complexo = 1j
print(type(complexo))

# decimal
from decimal import Decimal
s = Decimal(value='10', context=None)
print(s)
print(type(s))
```
