# _ _ builtins _ _
* São funcões pré importarados, ou sejam, já vem diretamente dentro do python, as mais famosas são o print e o input, mas tem muitas outras
* dir() serve para listar todos os membros do escopo atual(sem parametros) ou de um determinado objeto
  * Ele exibe funções tanto que nós programadores criamos, quanto as funções que já vem no python que podemos usar ou que já usamos
* help(valor) nos mostrar ajuda sobre como usar algo

```python
print(dir(), end=' ')
```

```python
print(dir(__builtins__), end=' ')
```

```python
help()
```

```python
help(print)
```
