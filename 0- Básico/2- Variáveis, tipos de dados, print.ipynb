# Variáveis
* Espaço na memória do computador utilizado para guardar dados, podendo ser alterado no decorrer do programa
  * Sobreescrição de variáveis: isso é colocar um valor por cima de outro valor na variável, sendo que o último valor é o que fica

# Regras para nomeação de variáveis
* É essencial a variável ter um nome pequeno
* Somente letras, números e o caractere undescore(_) podem ser usados
* O nome não pode começar com número
* Python é case sensitive, aonde, maiúsculas é diferente de minúsculas, ou seja, a variável Python é diferente da variável python
* Iniciar as variáveis com minúsculas é uma convensão da linguagem

# Constantes
* É um tipo de variável que não pode ser alterada

# print
* função que serve para exibir dados

# Atribuição
* =(igual) símbolo atribui o valor a esquerda ao valor que está a direita


```python
# variável
a = 10
print(a)

a = 5
print(a)

# Constante
A = 55
print(A)
```

# Tipos de dados(Básicos)
* int -> vem de inteiro, ou seja, só aceita valores inteiros, podendo tanto ser negativos quanto positivos
* float -> números fluatuantes, ou seja, números com vírgulas(ponto na programação)
* str -> strings, são um ou uma sequência de caracteres, e são definidos, por aspas simples ou compostas
* bol -> valores boleanos, são valores, que só aceitam True ou False


# Tipos de Dados(Coleção)
* list(lista)
* dict(dicionário)
* NoneType(sem tipo, None)


* type(valor) o type mostra de tipo que os valores são

```python
# int, float, str, bol
print(10, 10.5, 'World', True)
print(type(4e5)) # float(notação cientifica)
print(type(10.5))
print(type(10))
print(type('World'))
print(type(True))
print()

# list, dict, NoneType
lista = [10, 10, 10]
dicionario = {'a':1, 'b':2, 'c': 3}
none = None
print(lista, dicionario, none)
print(type(lista))
print(type(dicionario))
print(type(none))
```

# Operadores Arimtméticos(matemáticos)
* Obs.: Todos os símbolos aqui apresentados são sem Parênteses
  * parenteses serve para atribuir uma ordem de prioridade, ou seja, o que estiver dentro de parêntesis irá ser resolvido primeiro
* (+) Soma ou concatenação
* (-) subtração
* (*) Multiplicação
* (/) Divisão
* (//) Divisão de inteiros
* (**) exponenciação
* (%) módulo da divisão de inteiros(resto)
* (=) atribuição, coloca o resultado da expressão a direita na identificação(variável) a esquerda

```python
# atribuição
a = 10
b = 5

print('Soma:', a + b)
print('Subtração:', a - b)
print('Multiplicação:', a * b)
print('Divisão:', a / b)
print('Divisão de inteiros:', a // b)
print('Exponenciação:', a ** b)
print('Resto da divisão:', a % b)
```
