# Operadores Booleanos
* Aceitam apenas os valores True e False
* Os operadores booleanos são, and(e), or(ou), not(não) e or exclusive(ou exclusivo)

  * and -> é verdadeiro apenas se ambas as intruções forem verdadeiras
  * or -> é verdadeiro se ao menos uma das instruções for verdadeira
  * not -> inverte os valores, False torna-se True e vice-versa
  * or exclusive -> só é verdadeiro se pelo menos uma das instruções for falsa
    * and, or -> são operadores binários, ou seja, precisam de dois valores para compararem
    * o not atua apenas sobre uma expressão
    * posso mistura os valores boleanos e organizalos por parentesis
    
# Operadores de Comparação
* == igual
* != Diferente
* &lt; menor que
* &gt; maior que
* <= menor ou igual a 
* &gt;= maior ou igual a 

```python
print('Tabela verdade and: ')
print('True and True:', True and True)
print('True and False:', True and False)
print('False and True:', False and True)
print('False and False:', False and True)
print()

print('Tabela verdade or: ')
print('True or True:', True or True)
print('True or False:', True or False)
print('False or True:', False or True)
print('False or False:',False or False)
print()

print('Tabela verdade not: ')
print('not True:', not True)
print('not False: ', not False)
```

# Elementos do controle de fluxo
* **condição** São o mesmo que expressões, uma condição é somente um nome mais especifico no contexo das instruções de controle de fluxo, básicamente, valores que definem se vai acontecer caso algo seja True
* **Blocos de Código**  são básicamente linhas de código alinhadas
  * os blocos começam no local em que identação aumenta
  * os blocos podem conter outros blocos
  * os blocos terminam no local em que a identação se reduz a zero ou na identação do bloco que contém
* **Execução do programa** é um termo atual que se refere à instrução atual sendo executada
* **instruções de controle de fluxo** são as decisões que o programa irá tomar

# if...elif...else(são instruções de controle de fluxo)
* **if(se)** será executado apenas se sua condição for verdadeira
* **else(se não)** será executada apenas se nenhum condição for verdadeira
  * o else não é obrigátorio
* elif(se não se) só é executado se sua condição for verdadeira
  * o elif vem depois do if e não é obrigatório

```python
# if
a = 2
if a == 2:
  print('Igual')

# if...else
if a == 3:
   print('igual')
else:
   print('Diferente')


# if...elif...
if a == 3:
  print('Igual')
elif a == 2:
  print('Same Languages')



# if...elif...else
if a == '2':
  print(type(a))
elif a == 3:
  print('Valor?')
else:
  print('Valores diferentes')
```

# Os comandos in e not in
* in -> serve para verificar se determinado elemento está em algum lugar
* not in -> faz o inverso do in

```python
l = [1, 2, 3]
if 3 in l:
  print(True)
else:
  print(False)


if 5 not in l:
  print(True)
else:
  print(False)
```
