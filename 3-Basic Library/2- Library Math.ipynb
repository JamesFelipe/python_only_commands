# Biblioteca Math
* Assim como o nome diz, trata-se de uma biblioteca com propriedades matemáticas
* math.cos(valor) -> retorna o cosseno em radianos
* math.sin(valor) -> retorna o seno em radianos
* math.tan(valor) -> retorna a tangente em radianos
* math.pi -> retorna o valor de PI


```python
import math

print(math.cos(45))
print(math.pi)
```
