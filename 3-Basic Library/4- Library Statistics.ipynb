# Biblioteca statistics(Estatística)
* Trata de cáculos estatísticos básicos
* statistics.mean(coleção) -> tirá a média dos valores
* statistics.median(coleção) -> tirá a mediana dos valores
* statistics.variance(coleção) -> pega a variância dos dados

```python
import statistics

dados = list(range(1, 9))
print(statistics.mean(dados))
print(statistics.median(dados))
print(statistics.variance(dados))
```
