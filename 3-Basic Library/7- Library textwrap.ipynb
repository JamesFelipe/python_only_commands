# biblioteca textwrap
* serve para formata parágrafos de texto para que caibam em uma dada largura de tela
* textwrap.fill(variável, width=tamanho) -> é quem de fato adequa a largura

```python
import textwrap
python = '''python melhor que ... python melhor que ...python melhor que ...python melhor que ...
python melhor que ... python melhor que ...python melhor que ...python melhor que ...
python melhor que ... python melhor que ...python melhor que ...python melhor que as drogas :)'''
print(python)
print('\n')

print(textwrap.fill(python, width=68))

```
