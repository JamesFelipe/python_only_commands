# Biblioteca random(valores aleatórios)
* Serve para tratar ou gerar valores aleatórios
* random.choice(lista) -> serve para escolher uma valor aleátorio da lista
* random.sample(valor, valores_a_ser_exibido) -> Exibirá os valores de maneira aleatória
* random.random() -> valor aleatório de entre zero e um
* random.randrage(valor) -> gera valores aleatórios inteiros


```python
import random

animais = ['Leão', 'Jamanta', 'Cachorro']
print(random.choice(animais))
print()
```

```python
print(random.sample(range(100), 10))
```

```python
print(random.random())
```

```python
print(random.randrange(6))
```
